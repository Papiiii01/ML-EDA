```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Cleaned_GenZ_DatingApp_Data.csv")
```

Q1: Is there gender bias in dating app usage?

Gender Bias in Dating App Usage

The dataset has a fairly balanced gender distribution:
- Female: 35.2%
- Male: 34.2%
- Non-binary: 30.6%

There is no strong gender bias in representation, but the non-binary group is slightly underrepresented compared to male and female users.

Q2: Are certain demographics overrepresented in the dataset?

- Age: The most common ages are 24 (14.8%), 23 (14.6%), and 19 (13.4%), meaning young adults in their early 20s dominate the dataset.
- Location: Users are concentrated in Mumbai (16.4%), Delhi (15.2%), Kolkata (15.2%), Bangalore (13.8%), and Pune (13.4%), suggesting urban bias.

Q3: Is there a missing group of users that should have been included?

107 users (about 21.4%) did not specify a Primary App, indicating possible missing data for those who do not actively use dating apps or prefer alternatives not listed.


Q4: Can we trust this dataset for making general claims about Gen-Z dating in India?

- The dataset covers 8 unique age groups and 7 locations, but it is skewed towards urban Gen-Z users in major cities.
- It may not fully represent Gen-Z dating trends in smaller towns, rural areas, or less tech-savvy populations.

Activity 1: Use histograms and density plots to compare distributions across demographic groups.

```python
data["Daily_Usage_Time"] = data["Daily_Usage_Time"].replace({"30 minutes": 0.5, "1.5 hours": 1.5, "2 hours": 2, "3 hours": 3, "1 hour": 1})
```

```python
data.head()
```

```python
# Plot: Count of Users vs Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Gender', color="Yellow")
plt.title('Count of Users by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Users')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data["Age"], bins=10, kde=True, color="grey", edgecolor="black")
plt.title("Age Distribution of Users")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Activity 2: Identify overrepresented and underrepresented groups.

```python
# Location Distribution
plt.figure(figsize=(8, 5))
sns.countplot(y=data["Location"], color="Purple", edgecolor="black")
plt.title("Location Distribution of Users")
plt.xlabel("Count")
plt.ylabel("Location")
plt.show()
```

```python
# Density Plot: Daily usage time across locations
plt.figure(figsize=(10, 6))
sns.kdeplot(data, x='Daily_Usage_Time', hue='Location', fill=True)
plt.title('Daily App Usage Time Across Locations')
plt.xlabel('Daily Usage Time (Hours)')
plt.ylabel('Density')
plt.show()
```

Activity 2: Identify overrepresented and underrepresented groups.

```python
demographic_counts = {
    'Gender': data['Gender'].value_counts(),
    'Location': data['Location'].value_counts(),
    'Age Group': pd.cut(data['Age'], [18, 22, 26, 30, 35, 40]).value_counts()
}
demographic_counts
```

Discuss ethical considerations in data collection:

- Informed Consent: Clearly explain what data is being collected, how it will be used, and who will access it. Ensure participation is voluntary and can be withdrawn at any time. Obtain parental/guardian consent for minors or vulnerable groups.

- Privacy and Confidentiality: Protect personal data from unauthorized access. Use data encryption and secure storage methods. Anonymize or pseudonymize data where possible.

- Minimization of Data Collection: Collect only data that is necessary for the specific purpose. Avoid gathering excessive or irrelevant information. Regularly review data collection practices.

- Transparency and Accountability: Be open about how data is collected, stored, and used. Provide clear privacy policies and terms of use. Implement data governance practices and regular audits.

- Avoiding Harm: Ensure data collection does not harm individuals or communities. Avoid using data in ways that could lead to discrimination or exploitation. Handle sensitive data (health, financial) with extra care.

- Bias and Fairness: Design data collection tools to be inclusive and neutral. Regularly audit datasets for demographic or systemic biases. Be transparent about data limitations and potential biases.

- Legal Compliance: Adhere to data protection laws like GDPR, CCPA, and HIPAA. Provide mechanisms for individuals to access, correct, or delete their data. Ensure data processing aligns with legal requirements.

- Data Ownership and Control: Clarify who owns the data and how it will be used. Allow individuals to opt-out or delete their data if desired. Disclose any third-party data sharing.

- Data Security: Use encryption, firewalls, and secure access protocols. Regularly update security systems and perform vulnerability checks. Develop an incident response plan for data breaches.

- Ethical Use of Collected Data: Use data strictly for the purpose disclosed during collection. Avoid selling or misusing data without explicit consent. Consider the long-term societal impacts of data use.
